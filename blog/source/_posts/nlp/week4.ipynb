```python
def process_document(document: str) -> list:
    """
    pre-process a document and return a list of its terms
    str->list"""

    """ 
    pre-process steps:
    1. Document Coneversion
    2. Language identification
    3. tokenization (There could have the stop words remobal)
    4. Normalization
        1. case folding
        2. stemming
        3. lemmetization
    5. Spell correction
    """
    #step3 tokenization
    #
    tokens = word_tokenize(document)
    
    #case folding
    tokens = sorted(set(w.lower() for w in tokens))
    
    #get all the stop words in English
    stop_words = set(stopwords.words('english'))
    #stopwords removal
    tokens = [w for w in tokens if w not in stop_words]
    #remove no alphabetic tokens
    tokens = [w for w in tokens if w.isalpha()]
    #normalization
    #stemming
    #there are two stemmers in nltk
    porter = nltk.PorterStemmer()
    lancaster = nltk.LancasterStemmer()
    #using porter stemmer
    # tokens = [porter.stem(t) for t in tokens]
    #using Lancaster stermmer
    # tokens = [lancaster.stem(t) for t in tokens]
    return tokens



def document_frequency(term,documents):
    df  = len([True for t in documents if t['term'] == term])
    return df
def removing_key_from_documents(documents,key):
    for item in documents:
        del item[key]
    return documents
def flatten_list(documents):
    flat_list = []
    for sublist in documents:   
        for item in sublist:
            flat_list.append(item)
    return flat_list
def flatten_dictionary(documents):
    unique_terms = []
    flattened_documents = []
    for item in documents:
        t = item['term']
        if t not in unique_terms:
            unique_terms.append(t)
            flattened_documents.append(item)
    return flattened_documents
def doc_id_list(term,documents):
    posting_list = []
    for item in documents:
        if item['term'] == term:
            posting_list.append(item['Doc_id'])
    return posting_list

def dictionary_positing_lists(documents):
    for item in documents:
        item['posting_list'] = doc_id_list(item['term'],documents)
        item['doc_freq'] = document_frequency(item['term'],documents)
    #
    documents = flatten_dictionary(documents)
    #remove document id
    documents = removing_key_from_documents(documents,'Doc_id')

    dictionary = {}
    postings = {}
    for item in documents:
        dictionary[item['term']] = item['doc_freq']
        postings[item['term']] = item['posting_list']
      
    #return dictionary and postings
    return dictionary,postings
def dump(user_input,posting_lists):
    query_list = process_document(user_input)
    if len(query_list) == 1:
        if query_list[0] in posting_lists.keys():
            return posting_lists[query_list[0]]
    else:
        for term in query_list:
            if term in query_list:
                
            

        



    


test = [{'Doc_id': 1 , 'terms': ["jim","nick","helen","zed"]},{'Doc_id': 2 , 'terms': ["jim","nick","helen","zed"]},{'Doc_id': 0 , 'terms': ["jim","nick","helen","zed"]} ]


converted_documents = []
for document in test:
    converted_document = []
    for wd in document['terms']:
        tmp_dict = {}
        
        tmp_dict['term'] = wd
        tmp_dict['Doc_id'] = document['Doc_id']
        converted_document.append(tmp_dict)
    converted_documents.append(converted_document)

converted_documents = flatten_list(converted_documents)

converted_documents = sorted(converted_documents, key = lambda d : d['Doc_id'])
converted_documents = sorted(converted_documents, key = lambda d : d['term'])


dictionary,posting_lists = dictionary_positing_lists(converted_documents)


print(dictionary)
print(posting_lists)

print("Hello!")
terms_usrs_input = input("please tell us what do you wanna serarch")
print(dump(terms_usrs_input,posting_lists))



```

```python
import numpy as np
from numpy import array
from numpy.linalg import svd,matrix_rank
from numpy import dot
from numpy.linalg import norm
dtm = np.array([[1,1,0,0],
               [1,1,0,0],
               [0,0,1,1],
               [0,0,1,1],
               [1,1,0,1]])
u,d,v = svd(dtm)
#find how many columns are unique
k = matrix_rank(dtm)
U = u[:,:k]
V = v[:,:k]
D = np.diag(d)[:k,:k]

def cosine(a,b):
    return dot(a, b) / (norm(a) * norm(b))
#document vector and term vector
document_vector = np.matmul(U,D)
term_vector = np.matmul(V,D)
#get term vector for each term
v_matrix = term_vector[0]
v_factorization = term_vector[1]
v_nl = term_vector[2]
v_p = term_vector[3]

print(cosine(v_matrix, v_factorization))
print(cosine(v_matrix, v_nl))




```
